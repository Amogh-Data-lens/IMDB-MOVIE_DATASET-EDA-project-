```python
def rating(rating):
    if rating >= 7:
        return "Excellent"
    elif 6.0:
        return "Good"
    else:
        return "Average"
```

**22.Find unique values from Genre**

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
   list1.append(value.split(','))
```

```python


# Flatten the list using nested loops
flat_list = [item.strip() for sublist in list1 for item in sublist]

# Find unique values using set
unique_genres = set(flat_list)

# Convert the set back to a list if needed
unique_genres_list = list(unique_genres)

# Print the unique genres

print("Unique Genres:", unique_genres_list)
```

**23.How many films of each genre were made?**

```python
from collections import Counter

# Assuming 'data' is our DataFrame with a 'Genre' column
genre_counts = Counter()

for value in data['Genre']:
    genres = value.split(',')
    genre_counts.update(genres)

# Print the count of each genre
for genre, count in genre_counts.items():
    print(f"{genre}: {count} films")
```
